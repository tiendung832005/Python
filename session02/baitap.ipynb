```python
#Bài 4: Tính tổng các số từ 1 đến N
N = int(input("Nhap N = "))

# Khởi tạo biến tổng
tong = 0

# Sử dụng vòng lặp để tính tổng từ 1 đến N
for i in range(1, N + 1):
    tong += i

# In kết quả
print("Tong =", tong)
```

```python
#Bài 5: Kiểm tra số đối xứng (palindrome)
n = int(input("Nhap mot so nguyen duong: "))

# Chuyển số thành chuỗi để dễ kiểm tra
chuoi_so = str(n)

# Kiểm tra đối xứng bằng cách so sánh chuỗi với chuỗi đảo ngược
if chuoi_so == chuoi_so[::-1]:
    print(f"{n} la so doi xung")
else:
    print(f"{n} khong phai la so doi xung")
```

```python
#Bài 6: Tính giai thừa n!
n = int(input("Nhap mot so nguyen duong n: "))

# Khởi tạo giá trị giai thừa
giai_thua = 1

# Tính giai thừa bằng vòng lặp
for i in range(1, n + 1):
    giai_thua *= i

# In kết quả
print(f"{n}! = {giai_thua}")
```

```python
#Bài 7: Tìm số nguyên tố trong khoảng [a, b]

def kiem_tra_so_nguyen_to(n):
    """Hàm kiểm tra số nguyên tố"""
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Nhập hai số nguyên a và b
a = int(input("Nhap so a: "))
b = int(input("Nhap so b: "))

# Tìm và in các số nguyên tố trong khoảng [a, b]
print(f"Cac so nguyen to trong khoang [{a}, {b}]:")
co_so_nguyen_to = False

for num in range(a, b + 1):
    if kiem_tra_so_nguyen_to(num):
        print(num, end=" ")
        co_so_nguyen_to = True

if not co_so_nguyen_to:
    print("Khong co so nguyen to nao trong khoang nay")
else:
    print()  # Xuống dòng
```

```python
#Bài 8: Kiểm tra 3 cạnh tam giác

# Nhập 3 số nguyên đại diện cho 3 cạnh tam giác
a = int(input("Nhap canh a: "))
b = int(input("Nhap canh b: "))
c = int(input("Nhap canh c: "))

# Kiểm tra điều kiện tam giác
# Điều kiện: tổng 2 cạnh bất kỳ phải lớn hơn cạnh thứ 3
if (a + b > c) and (a + c > b) and (b + c > a):
    print("La 3 canh tam giac")
else:
    print("Khong phai 3 canh tam giac")
```
