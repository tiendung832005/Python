```python
#B√†i 2:
import csv
import os

def bai2():
    csv_filename = "students.csv"
    if not os.path.exists(csv_filename):
        with open(csv_filename, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(['name', 'age', 'grade'])  
            writer.writerow(['Nguy·ªÖn VƒÉn A', '20', '8.5'])
            writer.writerow(['Tr·∫ßn Th·ªã B', '19', '6.0'])
            writer.writerow(['L√™ VƒÉn C', '21', '7.5'])
            writer.writerow(['Ph·∫°m Th·ªã D', '20', '9.0'])
            writer.writerow(['Ho√†ng VƒÉn E', '22', '5.5'])
        print(f"ƒê√£ t·∫°o file m·∫´u {csv_filename}")
    
    # ƒê·ªçc d·ªØ li·ªáu t·ª´ file CSV
    students = []
    with open(csv_filename, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        header = next(reader) 
        for row in reader:
            students.append({
                'name': row[0],
                'age': int(row[1]),
                'grade': float(row[2])
            })
    
    print("=== T·∫•t c·∫£ h·ªçc sinh ===")
    for student in students:
        print(f"T√™n: {student['name']}, Tu·ªïi: {student['age']}, ƒêi·ªÉm: {student['grade']}")
    
    
    high_grade_students = [s for s in students if s['grade'] >= 7.0]
    
    print("\n=== H·ªçc sinh c√≥ ƒëi·ªÉm >= 7 ===")
    for student in high_grade_students:
        print(f"T√™n: {student['name']}, Tu·ªïi: {student['age']}, ƒêi·ªÉm: {student['grade']}")
    
    output_filename = "high_grade_students.csv"
    with open(output_filename, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['name', 'age', 'grade'])  
        for student in high_grade_students:
            writer.writerow([student['name'], student['age'], student['grade']])
    
    print(f"\nƒê√£ ghi {len(high_grade_students)} h·ªçc sinh c√≥ ƒëi·ªÉm cao v√†o file {output_filename}")

bai2()
```

```python
#B√†i 3
import json
import os

def bai3():
    json_filename = "user_data.json"
    
    if not os.path.exists(json_filename):
        sample_data = [
            {"id": 1, "name": "Nguy·ªÖn VƒÉn A", "age": 25},
            {"id": 2, "name": "Tr·∫ßn Th·ªã B", "age": 30},
            {"id": 3, "name": "L√™ VƒÉn C", "age": 28},
            {"id": 4, "name": "Ph·∫°m Th·ªã D", "age": 22}
        ]
        with open(json_filename, 'w', encoding='utf-8') as f:
            json.dump(sample_data, f, ensure_ascii=False, indent=2)
        print(f"ƒê√£ t·∫°o file m·∫´u {json_filename}")
    
    # ƒê·ªçc n·ªôi dung t·ª´ file JSON
    with open(json_filename, 'r', encoding='utf-8') as f:
        users = json.load(f)
    
    print("=== T·∫•t c·∫£ ng∆∞·ªùi d√πng ===")
    for user in users:
        print(f"ID: {user['id']}, T√™n: {user['name']}, Tu·ªïi: {user['age']}")
    
    target_user = None
    for user in users:
        if user['id'] == 3:
            target_user = user
            break
    
    if target_user:
        print(f"\n=== Ng∆∞·ªùi d√πng c√≥ ID = 3 ===")
        print(f"T√™n: {target_user['name']}")
        print(f"Tu·ªïi: {target_user['age']}")
    else:
        print("\nKh√¥ng t√¨m th·∫•y ng∆∞·ªùi d√πng c√≥ ID = 3")
    
    print("\n=== Th√™m ng∆∞·ªùi d√πng m·ªõi ===")
    new_id = max([user['id'] for user in users]) + 1
    new_name = input("Nh·∫≠p t√™n ng∆∞·ªùi d√πng m·ªõi: ")
    new_age = int(input("Nh·∫≠p tu·ªïi: "))
    
    new_user = {
        "id": new_id,
        "name": new_name,
        "age": new_age
    }
    
    users.append(new_user)
    
    with open(json_filename, 'w', encoding='utf-8') as f:
        json.dump(users, f, ensure_ascii=False, indent=2)
    
    print(f"ƒê√£ th√™m ng∆∞·ªùi d√πng m·ªõi v√† c·∫≠p nh·∫≠t file {json_filename}")
    
    print("\n=== Danh s√°ch sau khi th√™m ===")
    for user in users:
        print(f"ID: {user['id']}, T√™n: {user['name']}, Tu·ªïi: {user['age']}")

bai3()
```

```python
#B√†i 5
import os

def bai5():
    filename = "sample.txt"
    
    if not os.path.exists(filename):
        sample_content = """Python l√† m·ªôt ng√¥n ng·ªØ l·∫≠p tr√¨nh b·∫≠c cao.
Python ƒë∆∞·ª£c s·ª≠ d·ª•ng r·ªông r√£i trong nhi·ªÅu lƒ©nh v·ª±c kh√°c nhau.
Vi·ªác h·ªçc Python r·∫•t th√∫ v·ªã v√† b·ªï √≠ch.
Python c√≥ c√∫ ph√°p ƒë∆°n gi·∫£n v√† d·ªÖ hi·ªÉu."""
        
        with open(filename, 'w', encoding='utf-8') as f:
            f.write(sample_content)
        print(f"ƒê√£ t·∫°o file m·∫´u {filename}")
    
    print("=== ƒê·ªçc v√† ƒë·∫øm s·ªë t·ª´ trong file ===")
    
    with open(filename, 'r', encoding='utf-8') as f:
        content = f.read()
    
    print("N·ªôi dung file:")
    print(content)
    print("-" * 50)
    
    
    import re
    
    words = re.findall(r'\b\w+\b', content.lower())
    word_count = len(words)
    
    print(f"T·ªïng s·ªë t·ª´ trong file: {word_count}")
    
    lines = content.strip().split('\n')
    line_count = len(lines)
    print(f"S·ªë d√≤ng trong file: {line_count}")
    
    word_frequency = {}
    for word in words:
        word_frequency[word] = word_frequency.get(word, 0) + 1
    
    print("\n=== T·∫ßn su·∫•t xu·∫•t hi·ªán c·ªßa c√°c t·ª´ ===")
    sorted_words = sorted(word_frequency.items(), key=lambda x: x[1], reverse=True)
    for word, count in sorted_words:
        if count > 1:  
            print(f"'{word}': {count} l·∫ßn")

bai5()
```

```python
#B√†i 6
import json
import os

def bai6():
    json_filename = "inventory.json"
    
    if not os.path.exists(json_filename):
        sample_inventory = [
            {"product_id": 1, "name": "Laptop Dell", "quantity": 10, "price": 15000000},
            {"product_id": 2, "name": "Mouse Logitech", "quantity": 25, "price": 500000},
            {"product_id": 3, "name": "Keyboard Mechanical", "quantity": 15, "price": 1200000},
            {"product_id": 4, "name": "Monitor Samsung", "quantity": 8, "price": 5000000}
        ]
        
        with open(json_filename, 'w', encoding='utf-8') as f:
            json.dump(sample_inventory, f, ensure_ascii=False, indent=2)
        print(f"ƒê√£ t·∫°o file m·∫´u {json_filename}")
    
    print("=== Qu·∫£n l√Ω kho h√†ng ===")
    
    # ƒê·ªçc d·ªØ li·ªáu t·ª´ file JSON
    with open(json_filename, 'r', encoding='utf-8') as f:
        inventory = json.load(f)
    
    print("Danh s√°ch s·∫£n ph·∫©m hi·ªán t·∫°i:")
    print("-" * 80)
    print(f"{'ID':<5} {'T√™n s·∫£n ph·∫©m':<20} {'S·ªë l∆∞·ª£ng':<10} {'Gi√° (VNƒê)':<15}")
    print("-" * 80)
    
    for product in inventory:
        print(f"{product['product_id']:<5} {product['name']:<20} {product['quantity']:<10} {product['price']:,<15}")
    
    print("\n=== Th√™m s·∫£n ph·∫©m m·ªõi ===")
    new_id = max([product['product_id'] for product in inventory]) + 1
    
    new_name = input("Nh·∫≠p t√™n s·∫£n ph·∫©m: ")
    new_quantity = int(input("Nh·∫≠p s·ªë l∆∞·ª£ng: "))
    new_price = float(input("Nh·∫≠p gi√° (VNƒê): "))
    
    new_product = {
        "product_id": new_id,
        "name": new_name,
        "quantity": new_quantity,
        "price": new_price
    }
    
    inventory.append(new_product)
    
    with open(json_filename, 'w', encoding='utf-8') as f:
        json.dump(inventory, f, ensure_ascii=False, indent=2)
    
    print(f"\nƒê√£ th√™m s·∫£n ph·∫©m m·ªõi v√† c·∫≠p nh·∫≠t file {json_filename}")
    
    print("\nDanh s√°ch s·∫£n ph·∫©m sau khi th√™m:")
    print("-" * 80)
    print(f"{'ID':<5} {'T√™n s·∫£n ph·∫©m':<20} {'S·ªë l∆∞·ª£ng':<10} {'Gi√° (VNƒê)':<15}")
    print("-" * 80)
    
    for product in inventory:
        print(f"{product['product_id']:<5} {product['name']:<20} {product['quantity']:<10} {product['price']:,<15}")
    
    total_value = sum(product['quantity'] * product['price'] for product in inventory)
    print(f"\nT·ªïng gi√° tr·ªã kho h√†ng: {total_value:,} VNƒê")

bai6()
```

```python
#B√†i 7
from pathlib import Path

def bai7():
    print("=== Duy·ªát th∆∞ m·ª•c t√¨m file .txt b·∫±ng pathlib ===")
    
    sample_files = ["test1.txt", "test2.txt", "document.txt"]
    for filename in sample_files:
        if not Path(filename).exists():
            with open(filename, 'w', encoding='utf-8') as f:
                f.write(f"N·ªôi dung m·∫´u c·ªßa file {filename}\n")
    print("ƒê√£ t·∫°o c√°c file .txt m·∫´u ƒë·ªÉ test")
    
    current_dir = Path.cwd()
    print(f"Th∆∞ m·ª•c hi·ªán t·∫°i: {current_dir}")
    
    print("\n=== T·∫•t c·∫£ file v√† th∆∞ m·ª•c ===")
    all_items = list(current_dir.iterdir())
    
    for item in sorted(all_items):
        if item.is_file():
            print(f"üìÑ File: {item.name} (K√≠ch th∆∞·ªõc: {item.stat().st_size} bytes)")
        elif item.is_dir():
            print(f"üìÅ Th∆∞ m·ª•c: {item.name}")
    
    print("\n=== File c√≥ ƒëu√¥i .txt ===")
    txt_files = []
    
    for item in current_dir.iterdir():
        if item.is_file() and item.suffix.lower() == '.txt':
            txt_files.append(item)
    
    if txt_files:
        print(f"T√¨m th·∫•y {len(txt_files)} file .txt:")
        for txt_file in sorted(txt_files):
            file_size = txt_file.stat().st_size
            print(f"  ‚Ä¢ {txt_file.name} (K√≠ch th∆∞·ªõc: {file_size} bytes)")
            
            try:
                with open(txt_file, 'r', encoding='utf-8') as f:
                    first_line = f.readline().strip()
                    if first_line:
                        print(f"    N·ªôi dung ƒë·∫ßu: {first_line}")
            except Exception as e:
                print(f"    Kh√¥ng th·ªÉ ƒë·ªçc file: {e}")
    else:
        print("Kh√¥ng t√¨m th·∫•y file .txt n√†o trong th∆∞ m·ª•c hi·ªán t·∫°i.")
    
    print("\n=== Th·ªëng k√™ theo lo·∫°i file ===")
    file_extensions = {}
    
    for item in current_dir.iterdir():
        if item.is_file():
            ext = item.suffix.lower() if item.suffix else '(kh√¥ng c√≥ ƒëu√¥i)'
            file_extensions[ext] = file_extensions.get(ext, 0) + 1
    
    for ext, count in sorted(file_extensions.items()):
        print(f"  {ext}: {count} file(s)")

bai7()
```

```python
#B√†i 8
import datetime
import os

def bai8():
    log_filename = "log.txt"
    
    if not os.path.exists(log_filename):
        now = datetime.datetime.now()
        today = now.date()
        yesterday = today - datetime.timedelta(days=1)
        tomorrow = today + datetime.timedelta(days=1)
        
        sample_logs = [
            f"{yesterday} 08:30:15 - H·ªá th·ªëng kh·ªüi ƒë·ªông",
            f"{yesterday} 09:15:22 - Ng∆∞·ªùi d√πng ƒëƒÉng nh·∫≠p",
            f"{yesterday} 14:45:33 - L·ªói k·∫øt n·ªëi c∆° s·ªü d·ªØ li·ªáu",
            f"{today} 07:20:10 - H·ªá th·ªëng kh·ªüi ƒë·ªông",
            f"{today} 08:45:25 - Ng∆∞·ªùi d√πng Admin ƒëƒÉng nh·∫≠p",
            f"{today} 10:30:45 - Backup d·ªØ li·ªáu th√†nh c√¥ng",
            f"{today} 12:15:18 - C·∫£nh b√°o: Dung l∆∞·ª£ng ƒëƒ©a th·∫•p",
            f"{today} 15:22:33 - Ng∆∞·ªùi d√πng ƒëƒÉng xu·∫•t",
            f"{today} 16:45:12 - C·∫≠p nh·∫≠t ph·∫ßn m·ªÅm",
            f"{tomorrow} 06:00:00 - Scheduled maintenance"
        ]
        
        with open(log_filename, 'w', encoding='utf-8') as f:
            for log in sample_logs:
                f.write(log + '\n')
        
        print(f"ƒê√£ t·∫°o file log m·∫´u {log_filename}")
    
    print("=== ƒê·ªçc file log v√† l·ªçc theo ng√†y ===")
    
    today = datetime.date.today()
    print(f"Ng√†y h√¥m nay: {today}")
    
    with open(log_filename, 'r', encoding='utf-8') as f:
        log_lines = f.readlines()
    
    print(f"\nT·ªïng s·ªë d√≤ng log: {len(log_lines)}")
    print("\n=== T·∫•t c·∫£ c√°c d√≤ng log ===")
    for i, line in enumerate(log_lines, 1):
        print(f"{i:2d}. {line.strip()}")
    
    today_logs = []
    
    for line in log_lines:
        line = line.strip()
        if line:
            try:
                parts = line.split(' - ', 1)
                if len(parts) >= 2:
                    timestamp_str = parts[0].split()[0]  
                    
                    log_date = datetime.datetime.strptime(timestamp_str, '%Y-%m-%d').date()
                    
                    if log_date == today:
                        today_logs.append(line)
                        
            except ValueError:
                continue
    
    print(f"\n=== C√°c d√≤ng log c·ªßa ng√†y h√¥m nay ({today}) ===")
    if today_logs:
        for i, log in enumerate(today_logs, 1):
            print(f"{i:2d}. {log}")
        
        print(f"\nT√¨m th·∫•y {len(today_logs)} d√≤ng log c·ªßa ng√†y h√¥m nay.")
        
        print("\n=== Ph√¢n t√≠ch n·ªôi dung log h√¥m nay ===")
        keywords = ['kh·ªüi ƒë·ªông', 'ƒëƒÉng nh·∫≠p', 'ƒëƒÉng xu·∫•t', 'l·ªói', 'c·∫£nh b√°o', 'th√†nh c√¥ng']
        
        for keyword in keywords:
            count = sum(1 for log in today_logs if keyword.lower() in log.lower())
            if count > 0:
                print(f"  ‚Ä¢ {keyword.capitalize()}: {count} l·∫ßn")
                
    else:
        print("Kh√¥ng t√¨m th·∫•y d√≤ng log n√†o c·ªßa ng√†y h√¥m nay.")
        
    if today_logs:
        print(f"\n=== Th·ªëng k√™ th·ªùi gian ===")
        times = []
        for log in today_logs:
            try:
                timestamp_str = log.split(' - ')[0]
                time_str = timestamp_str.split()[1]  
                times.append(time_str)
            except:
                continue
        
        if times:
            print(f"Log ƒë·∫ßu ti√™n: {times[0]}")
            print(f"Log cu·ªëi c√πng: {times[-1]}")

bai8()
```
