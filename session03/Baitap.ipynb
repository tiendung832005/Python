```python
# Bài 1: Kiểm tra số nguyên tố
def is_prime(n):
    """
    Kiểm tra xem một số có phải là số nguyên tố hay không
    """
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Test hàm
test_numbers = [2, 3, 4, 5, 17, 25, 29, 100]
for num in test_numbers:
    result = is_prime(num)
    print(f"{num} {'là' if result else 'không phải là'} số nguyên tố")
```

```python
# Bài 2: Kiểm tra số nguyên tố (phương pháp khác)
def check_prime(number):
    """
    Kiểm tra số nguyên tố bằng cách đơn giản hơn
    """
    if number <= 1:
        return False
    if number == 2:
        return True
    if number % 2 == 0:
        return False
    
    for i in range(3, int(number**0.5) + 1, 2):
        if number % i == 0:
            return False
    return True

try:
    user_input = int(input("Nhập một số để kiểm tra: "))
    if check_prime(user_input):
        print(f"{user_input} là số nguyên tố")
    else:
        print(f"{user_input} không phải là số nguyên tố")
except ValueError:
    print("Vui lòng nhập một số nguyên hợp lệ")
```

```python
# Bài 3: Tính giai thừa
def factorial(n):
    """
    Tính giai thừa của số n
    n! = n * (n-1) * (n-2) * ... * 2 * 1
    """
    if n < 0:
        return "Giai thừa không xác định cho số âm"
    elif n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

test_values = [0, 1, 5, 7, 10]
for val in test_values:
    print(f"{val}! = {factorial(val)}")

def factorial_recursive(n):
    """Tính giai thừa bằng đệ quy"""
    if n < 0:
        return "Giai thừa không xác định cho số âm"
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n - 1)

print(f"\nSử dụng đệ quy: 5! = {factorial_recursive(5)}")
```

```python
# Bài 4: Tính diện tích hình tròn
def circle_area(radius):
    """
    Tính diện tích hình tròn
    Công thức: A = π * r²
    """
    pi = 3.14159
    if radius < 0:
        return "Bán kính không thể âm"
    return pi * radius * radius

test_radii = [1, 2.5, 5, 10, 0]
for r in test_radii:
    area = circle_area(r)
    print(f"Bán kính = {r}, Diện tích = {area}")

try:
    user_radius = float(input("\nNhập bán kính hình tròn: "))
    result = circle_area(user_radius)
    if isinstance(result, str):
        print(result)
    else:
        print(f"Diện tích hình tròn với bán kính {user_radius} là: {result:.2f}")
except ValueError:
    print("Vui lòng nhập một số hợp lệ")
```

```python
# Bài 5: Kiểm tra năm nhuận
def is_leap_year(year):
    """
    Kiểm tra năm nhuận
    Quy tắc:
    - Chia hết cho 4 AND (không chia hết cho 100 OR chia hết cho 400)
    """
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True 
            else:
                return False  
        else:
            return True 
    else:
        return False  


test_years = [2000, 2004, 2100, 2020, 2021, 1900, 2024]
for year in test_years:
    result = is_leap_year(year)
    print(f"Năm {year} {'là' if result else 'không phải là'} năm nhuận")


try:
    user_year = int(input("\nNhập một năm để kiểm tra: "))
    if is_leap_year(user_year):
        print(f"Năm {user_year} là năm nhuận")
    else:
        print(f"Năm {user_year} không phải là năm nhuận")
except ValueError:
    print("Vui lòng nhập một năm hợp lệ")
```

```python
# Bài 6: Tính tổng các chữ số
def sum_of_digits(number):
    """
    Tính tổng các chữ số trong một số nguyên
    Ví dụ: 12345 = 1 + 2 + 3 + 4 + 5 = 15
    """
    
    number = abs(number)
    total = 0
    
    while number > 0:
        digit = number % 10  
        total += digit    
        number //= 10       
    
    return total


def sum_of_digits_string(number):
    """Tính tổng các chữ số bằng cách chuyển thành chuỗi"""
    return sum(int(digit) for digit in str(abs(number)))


test_numbers = [12345, 999, 0, 7, -12345, 100]
for num in test_numbers:
    sum1 = sum_of_digits(num)
    sum2 = sum_of_digits_string(num)
    print(f"Số {num}: Tổng các chữ số = {sum1} (phương pháp 1), {sum2} (phương pháp 2)")


try:
    user_number = int(input("\nNhập một số để tính tổng các chữ số: "))
    result = sum_of_digits(user_number)
    print(f"Tổng các chữ số của {user_number} là: {result}")
except ValueError:
    print("Vui lòng nhập một số nguyên hợp lệ")
```

```python
# Bài 7: Đảo ngược một số nguyên
def reverse_number(number):
    """
    Đảo ngược một số nguyên
    Ví dụ: 12345 -> 54321
    """
    
    is_negative = number < 0
    number = abs(number)
    
    reversed_num = 0
    while number > 0:
        digit = number % 10     
        reversed_num = reversed_num * 10 + digit  
        number //= 10         
    
    
    return -reversed_num if is_negative else reversed_num


def reverse_number_string(number):
    """Đảo ngược số bằng cách chuyển thành chuỗi"""
    is_negative = number < 0
    number_str = str(abs(number))
    reversed_str = number_str[::-1]  
    reversed_num = int(reversed_str)
    return -reversed_num if is_negative else reversed_num


test_numbers = [12345, 1000, 7, -12345, 0, 120]
for num in test_numbers:
    rev1 = reverse_number(num)
    rev2 = reverse_number_string(num)
    print(f"Số gốc: {num}, Đảo ngược: {rev1} (phương pháp 1), {rev2} (phương pháp 2)")


try:
    user_input = int(input("\nNhập một số để đảo ngược: "))
    result = reverse_number(user_input)
    print(f"Số {user_input} sau khi đảo ngược là: {result}")
except ValueError:
    print("Vui lòng nhập một số nguyên hợp lệ")
```

```python
# Bài 8: Kiểm tra số hoàn hảo
def is_perfect_number(n):
    """
    Kiểm tra số hoàn hảo
    Số hoàn hảo là số bằng tổng các ước số dương nhỏ hơn nó
    Ví dụ: 6 = 1 + 2 + 3 (các ước của 6 nhỏ hơn 6)
    """
    if n <= 1:
        return False
    
    # Tìm tất cả các ước số 
    divisors_sum = 1  
    
    
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            divisors_sum += i
            
            if i != n // i:
                divisors_sum += n // i
    
    return divisors_sum == n


print("Kiểm tra các số hoàn hảo:")
test_numbers = [6, 28, 496, 8128, 12, 100]
for num in test_numbers:
    result = is_perfect_number(num)
    print(f"{num} {'là' if result else 'không phải là'} số hoàn hảo")


print("\nCác số hoàn hảo từ 1 đến 1000:")
perfect_numbers = []
for i in range(1, 1001):
    if is_perfect_number(i):
        perfect_numbers.append(i)
print(perfect_numbers)


print("\n" + "="*50)
try:
    user_number = int(input("Nhập một số nguyên để kiểm tra số hoàn hảo: "))
    if is_perfect_number(user_number):
        print(f"Số {user_number} là số hoàn hảo!")
        
       
        divisors = []
        for i in range(1, user_number):
            if user_number % i == 0:
                divisors.append(i)
        print(f"Các ước số của {user_number} (nhỏ hơn {user_number}): {divisors}")
        print(f"Tổng các ước số: {sum(divisors)} = {user_number}")
    else:
        print(f"Số {user_number} không phải là số hoàn hảo.")
        
except ValueError:
    print("Vui lòng nhập một số nguyên hợp lệ!")
```
