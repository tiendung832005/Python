```python
# Bài 1: Phân tích phân phối điểm thi
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

scores = np.random.normal(7, 1.5, 200)
scores = np.clip(scores, 0, 10)  # Giới hạn từ 0-10

plt.figure(figsize=(10, 6))

sns.histplot(scores, kde=True, bins=20, alpha=0.7, color='skyblue')

plt.title('Phân phối điểm thi của học sinh', fontsize=14, fontweight='bold')
plt.xlabel('Điểm số', fontsize=12)
plt.ylabel('Tần suất', fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Nhận xét:")
print(f"- Điểm trung bình: {scores.mean():.2f}")
print(f"- Độ lệch chuẩn: {scores.std():.2f}")
print("- Phân phối có dạng chuông, tập trung quanh điểm 7")
print("- Đa số học sinh có điểm từ 5.5 đến 8.5")
```

```python
# Bài 2: So sánh điểm các môn học
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
subjects = ['Toán'] * 100 + ['Lý'] * 100 + ['Hóa'] * 100
math_scores = np.random.normal(7.5, 1.2, 100)
physics_scores = np.random.normal(6.8, 1.5, 100)
chemistry_scores = np.random.normal(7.2, 1.0, 100)

scores = np.concatenate([math_scores, physics_scores, chemistry_scores])
scores = np.clip(scores, 0, 10)

df = pd.DataFrame({
    'subject': subjects,
    'score': scores
})

plt.figure(figsize=(10, 6))

sns.displot(df, x="score", hue="subject", kind="kde", fill=True, alpha=0.6)

plt.title('So sánh phân phối điểm các môn học', fontsize=14, fontweight='bold')
plt.xlabel('Điểm số', fontsize=12)
plt.ylabel('Mật độ', fontsize=12)

plt.tight_layout()
plt.show()

print("Nhận xét:")
print(f"- Toán có điểm trung bình cao nhất: {df[df['subject']=='Toán']['score'].mean():.2f}")
print(f"- Lý có điểm trung bình thấp nhất: {df[df['subject']=='Lý']['score'].mean():.2f}")
print(f"- Hóa có điểm trung bình: {df[df['subject']=='Hóa']['score'].mean():.2f}")
print("- Hóa có phân phối tập trung nhất, Lý có phân tán lớn nhất")
```

```python
# Bài 3: Mối quan hệ giữa số giờ học và điểm thi
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(123)
hours = np.random.uniform(1, 10, 150)
noise = np.random.normal(0, 0.8, 150)
score = hours * 1.2 + noise
score = np.clip(score, 0, 12)  # Giới hạn điểm

plt.figure(figsize=(10, 6))

sns.scatterplot(x=hours, y=score, alpha=0.7, s=60, color='coral')

plt.title('Mối quan hệ giữa số giờ học và điểm thi', fontsize=14, fontweight='bold')
plt.xlabel('Số giờ học', fontsize=12)
plt.ylabel('Điểm số', fontsize=12)
plt.grid(True, alpha=0.3)

z = np.polyfit(hours, score, 1)
p = np.poly1d(z)
plt.plot(hours, p(hours), "r--", alpha=0.8, linewidth=2, label=f'Đường hồi quy: y = {z[0]:.2f}x + {z[1]:.2f}')
plt.legend()

plt.tight_layout()
plt.show()
correlation = np.corrcoef(hours, score)[0,1]
print("Nhận xét:")
print(f"- Hệ số tương quan: {correlation:.3f}")
print("- Có mối quan hệ tương quan dương mạnh giữa số giờ học và điểm thi")
print("- Số giờ học tăng thì điểm thi có xu hướng tăng theo")
print("- Mỗi giờ học thêm có thể giúp tăng khoảng 1.2 điểm")
```

```python
# Bài 4: So sánh điểm trung bình các lớp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(456)
classes = ['A'] * 40 + ['B'] * 35 + ['C'] * 38
class_a_scores = np.random.normal(8.2, 1.0, 40)
class_b_scores = np.random.normal(7.1, 1.3, 35)
class_c_scores = np.random.normal(8.7, 0.8, 38)

scores = np.concatenate([class_a_scores, class_b_scores, class_c_scores])
scores = np.clip(scores, 0, 10)

df = pd.DataFrame({
    'class': classes,
    'score': scores
})

plt.figure(figsize=(8, 6))

sns.barplot(data=df, x='class', y='score', palette='viridis', ci=95)

plt.title('Điểm trung bình của các lớp', fontsize=14, fontweight='bold')
plt.xlabel('Lớp', fontsize=12)
plt.ylabel('Điểm trung bình', fontsize=12)
plt.grid(True, alpha=0.3, axis='y')

for i, cls in enumerate(['A', 'B', 'C']):
    mean_score = df[df['class']==cls]['score'].mean()
    plt.text(i, mean_score + 0.1, f'{mean_score:.2f}', 
             ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("Nhận xét:")
for cls in ['A', 'B', 'C']:
    mean_score = df[df['class']==cls]['score'].mean()
    print(f"- Lớp {cls}: điểm trung bình {mean_score:.2f}")

print("\n- Lớp C có điểm trung bình cao nhất")
print("- Lớp B có điểm trung bình thấp nhất")
print("- Chênh lệch giữa lớp cao nhất và thấp nhất là đáng kể (>1.5 điểm)")
```

```python
# Bài 5: So sánh phân phối điểm các ngành học
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(789)
majors = ['CNTT'] * 80 + ['Kinh Tế'] * 75 + ['Ngôn Ngữ'] * 85
cntt_scores = np.random.normal(7.8, 1.5, 80)
econ_scores = np.random.normal(8.1, 0.9, 75)  
lang_scores = np.random.normal(7.5, 1.8, 85)

scores = np.concatenate([cntt_scores, econ_scores, lang_scores])
scores = np.clip(scores, 0, 10)

df = pd.DataFrame({
    'major': majors,
    'score': scores
})

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='major', y='score', palette='Set2')
plt.title('Boxplot - Phân phối điểm các ngành', fontsize=12, fontweight='bold')
plt.ylabel('Điểm số', fontsize=11)
plt.xlabel('Ngành học', fontsize=11)

plt.subplot(1, 2, 2)
sns.violinplot(data=df, x='major', y='score', palette='Set2')
plt.title('Violinplot - Mật độ phân phối điểm', fontsize=12, fontweight='bold')
plt.ylabel('Điểm số', fontsize=11)
plt.xlabel('Ngành học', fontsize=11)

plt.tight_layout()
plt.show()

print("Nhận xét chi tiết:")
for major in ['CNTT', 'Kinh Tế', 'Ngôn Ngữ']:
    data = df[df['major']==major]['score']
    print(f"\n{major}:")
    print(f"  - Trung vị: {data.median():.2f}")
    print(f"  - Trung bình: {data.mean():.2f}")
    print(f"  - Độ lệch chuẩn: {data.std():.2f}")

print("\nKết luận:")
print("- Kinh Tế có điểm tập trung nhất (độ lệch chuẩn nhỏ)")
print("- Ngôn Ngữ có điểm phân tán rộng nhất")
print("- Kinh Tế có trung vị cao nhất")
print("- CNTT có phân phối cân đối nhất")
```

```python
# Bài 6: Phân tích xu hướng doanh thu theo thời gian
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(111)
days = list(range(1, 31))
base_sales = [day * 2 + 50 for day in days]  # Xu hướng tăng
noise = np.random.normal(0, 8, 30)
sales = [base + n for base, n in zip(base_sales, noise)]
sales = [max(0, s) for s in sales]  # Không âm

df = pd.DataFrame({
    'day': days,
    'sales': sales
})

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.lineplot(data=df, x='day', y='sales', marker='o', linewidth=2, markersize=6)
plt.title('Biến động doanh thu theo ngày', fontsize=12, fontweight='bold')
plt.xlabel('Ngày', fontsize=11)
plt.ylabel('Doanh thu (triệu đồng)', fontsize=11)
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
sns.regplot(data=df, x='day', y='sales', scatter_kws={'alpha':0.7}, line_kws={'color':'red'})
plt.title('Xu hướng doanh thu với đường hồi quy', fontsize=12, fontweight='bold')
plt.xlabel('Ngày', fontsize=11)
plt.ylabel('Doanh thu (triệu đồng)', fontsize=11)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

correlation = np.corrcoef(days, sales)[0,1]
slope = np.polyfit(days, sales, 1)[0]

print("Nhận xét về xu hướng doanh thu:")
print(f"- Hệ số tương quan: {correlation:.3f}")
print(f"- Mức tăng trung bình: {slope:.2f} triệu đồng/ngày")
print("- Doanh thu có xu hướng tăng rõ rệt theo thời gian")
print("- Biến động hàng ngày tương đối ổn định")
print("- Đường hồi quy thể hiện xu hướng tăng trưởng tích cực")
```

```python
# Bài 7: Dashboard mini - Phân tích đa chiều dữ liệu sinh viên
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(999)
n_students = 500

hours = np.random.uniform(1, 10, n_students)
noise = np.random.normal(0, 1.2, n_students)
score = hours * 0.8 + 2 + noise
score = np.clip(score, 0, 10)

majors = np.random.choice(['A', 'B', 'C'], n_students, p=[0.35, 0.30, 0.35])

df = pd.DataFrame({
    'hours': hours,
    'score': score,
    'major': majors
})

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['score'], bins=25, kde=True, color='lightblue', alpha=0.7)
plt.title('Phân phối điểm thi', fontsize=12, fontweight='bold')
plt.xlabel('Điểm số', fontsize=11)
plt.ylabel('Tần suất', fontsize=11)

plt.subplot(1, 3, 2)
sns.scatterplot(data=df, x='hours', y='score', alpha=0.6, s=40, color='green')
plt.title('Mối quan hệ giờ học - điểm thi', fontsize=12, fontweight='bold')
plt.xlabel('Số giờ học', fontsize=11)
plt.ylabel('Điểm số', fontsize=11)

z = np.polyfit(df['hours'], df['score'], 1)
p = np.poly1d(z)
plt.plot(df['hours'], p(df['hours']), "r--", alpha=0.8, linewidth=2)

plt.subplot(1, 3, 3)
sns.boxplot(data=df, x='major', y='score', palette='viridis')
plt.title('So sánh điểm theo ngành', fontsize=12, fontweight='bold')
plt.xlabel('Ngành', fontsize=11)
plt.ylabel('Điểm số', fontsize=11)

plt.tight_layout()
plt.show()

print("=== KẾT LUẬN TỔNG HỢP ===")
print("\n1. PHÂN PHỐI ĐIỂM:")
print(f"   - Điểm trung bình: {df['score'].mean():.2f}")
print(f"   - Phân phối gần chuẩn, tập trung quanh {df['score'].mean():.1f}")

print("\n2. MỐI QUAN HỆ GIỜ HỌC - ĐIỂM:")
correlation = df['hours'].corr(df['score'])
print(f"   - Tương quan: {correlation:.3f} (tương quan dương trung bình)")
print(f"   - Mỗi giờ học thêm tăng khoảng {z[0]:.2f} điểm")

print("\n3. SO SÁNH CÁC NGÀNH:")
for major in ['A', 'B', 'C']:
    avg = df[df['major']==major]['score'].mean()
    print(f"   - Ngành {major}: {avg:.2f} điểm")
print(f"   - Chênh lệch giữa ngành cao nhất và thấp nhất: {df.groupby('major')['score'].mean().max() - df.groupby('major')['score'].mean().min():.2f} điểm")
```

```python
# Bài 8: Mini Analytics - Phân tích doanh số bán hàng đa chiều
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(2024)
months = list(range(1, 13)) * 3  # 12 tháng x 3 nhóm sản phẩm
categories = ['Shoes'] * 12 + ['Clothes'] * 12 + ['Bags'] * 12

shoes_revenue = [month * 5 + np.random.normal(60, 15) for month in range(1, 13)]
clothes_revenue = [month * 3 + np.random.normal(80, 10) for month in range(1, 13)]
bags_revenue = [month * 2 + np.random.normal(45, 20) for month in range(1, 13)]

revenue = shoes_revenue + clothes_revenue + bags_revenue
revenue = [max(10, r) for r in revenue]  # Không nhỏ hơn 10

df = pd.DataFrame({
    'month': months,
    'revenue': revenue,
    'category': categories
})

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.lineplot(data=df, x='month', y='revenue', hue='category', marker='o', linewidth=2.5)
plt.title('Xu hướng doanh thu theo tháng', fontsize=12, fontweight='bold')
plt.xlabel('Tháng', fontsize=11)
plt.ylabel('Doanh thu (triệu)', fontsize=11)
plt.grid(True, alpha=0.3)
plt.legend(title='Nhóm sản phẩm')

plt.subplot(1, 3, 2)
sns.violinplot(data=df, x='category', y='revenue', palette='Set2')
plt.title('Phân phối doanh thu các nhóm', fontsize=12, fontweight='bold')
plt.xlabel('Nhóm sản phẩm', fontsize=11)
plt.ylabel('Doanh thu (triệu)', fontsize=11)

plt.subplot(1, 3, 3)
sns.regplot(data=df, x='month', y='revenue', scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title('Xu hướng chung với đường hồi quy', fontsize=12, fontweight='bold')
plt.xlabel('Tháng', fontsize=11)
plt.ylabel('Doanh thu (triệu)', fontsize=11)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("=== PHÂN TÍCH TỔNG QUAN ===")

for cat in ['Shoes', 'Clothes', 'Bags']:
    cat_data = df[df['category']==cat]
    mean_rev = cat_data['revenue'].mean()
    std_rev = cat_data['revenue'].std()
    trend = np.polyfit(cat_data['month'], cat_data['revenue'], 1)[0]
    print(f"{cat}: TB={mean_rev:.1f}, Biến động={std_rev:.1f}, Xu hướng={trend:.2f}/tháng")

print(f"\nKẾT LUẬN:")
print(f"• Clothes có doanh thu ổn định nhất với biến động thấp và xu hướng tăng đều.")
print(f"• Shoes có xu hướng tăng mạnh nhất theo thời gian (+{np.polyfit(df[df['category']=='Shoes']['month'], df[df['category']=='Shoes']['revenue'], 1)[0]:.1f}/tháng).")
print(f"• Bags có doanh thu thấp nhất và biến động lớn, cho thấy tính không ổn định.")
print(f"• Xu hướng chung của thị trường là tăng trưởng tích cực qua các tháng.")
print(f"• Có sự khác biệt rõ rệt giữa ba nhóm, với Clothes dẫn đầu về doanh thu trung bình.")
```
