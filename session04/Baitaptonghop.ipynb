```python
# B√†i t·∫≠p t·ªïng h·ª£p: Qu·∫£n l√Ω s·∫£n ph·∫©m c·ª≠a h√†ng

# Danh s√°ch l∆∞u tr·ªØ s·∫£n ph·∫©m
products = []

def display_menu():
    """Hi·ªÉn th·ªã menu l·ª±a ch·ªçn"""
    print("\n" + "="*50)
    print("         QU·∫¢N L√ù S·∫¢N PH·∫®M C·ª¨A H√ÄNG")
    print("="*50)
    print("1. Th√™m s·∫£n ph·∫©m m·ªõi")
    print("2. Hi·ªÉn th·ªã danh s√°ch s·∫£n ph·∫©m") 
    print("3. T√¨m ki·∫øm s·∫£n ph·∫©m theo m√£ s·∫£n ph·∫©m")
    print("4. C·∫≠p nh·∫≠t th√¥ng tin s·∫£n ph·∫©m")
    print("5. X√≥a s·∫£n ph·∫©m theo m√£ s·∫£n ph·∫©m")
    print("0. Tho√°t ch∆∞∆°ng tr√¨nh")
    print("="*50)

def find_product_by_id(product_id):
    """T√¨m s·∫£n ph·∫©m theo m√£ s·∫£n ph·∫©m, tr·∫£ v·ªÅ index n·∫øu t√¨m th·∫•y"""
    for i, product in enumerate(products):
        if product[0] == product_id:
            return i
    return -1

def add_product():
    """Th√™m s·∫£n ph·∫©m m·ªõi"""
    print("\n--- TH√äM S·∫¢N PH·∫®M M·ªöI ---")
    
    # Nh·∫≠p th√¥ng tin s·∫£n ph·∫©m
    product_id = input("Nh·∫≠p m√£ s·∫£n ph·∫©m: ").strip()
    
    # Ki·ªÉm tra m√£ s·∫£n ph·∫©m c√≥ tr√πng l·∫∑p kh√¥ng
    if find_product_by_id(product_id) != -1:
        print(f" L·ªói: M√£ s·∫£n ph·∫©m '{product_id}' ƒë√£ t·ªìn t·∫°i!")
        return
    
    name = input("Nh·∫≠p t√™n s·∫£n ph·∫©m: ").strip()
    
    try:
        quantity = int(input("Nh·∫≠p s·ªë l∆∞·ª£ng: "))
        price = float(input("Nh·∫≠p gi√°: "))
        
        if quantity < 0 or price < 0:
            print(" L·ªói: S·ªë l∆∞·ª£ng v√† gi√° kh√¥ng th·ªÉ √¢m!")
            return
            
        # T·∫°o tuple s·∫£n ph·∫©m v√† th√™m v√†o danh s√°ch
        product = (product_id, name, quantity, price)
        products.append(product)
        print(f" ƒê√£ th√™m s·∫£n ph·∫©m '{name}' th√†nh c√¥ng!")
        
    except ValueError:
        print(" L·ªói: Vui l√≤ng nh·∫≠p s·ªë h·ª£p l·ªá cho s·ªë l∆∞·ª£ng v√† gi√°!")

def display_products():
    """Hi·ªÉn th·ªã danh s√°ch t·∫•t c·∫£ s·∫£n ph·∫©m"""
    print("\n--- DANH S√ÅCH S·∫¢N PH·∫®M ---")
    
    if not products:
        print("üìã Danh s√°ch s·∫£n ph·∫©m tr·ªëng!")
        return
    
    print(f"{'STT':<5}{'M√£ SP':<10}{'T√™n s·∫£n ph·∫©m':<25}{'S·ªë l∆∞·ª£ng':<12}{'Gi√°':<15}")
    print("-" * 70)
    
    for i, (product_id, name, quantity, price) in enumerate(products, 1):
        print(f"{i:<5}{product_id:<10}{name:<25}{quantity:<12}{price:,.0f} VND")
    
    print(f"\nT·ªïng s·ªë s·∫£n ph·∫©m: {len(products)}")

def search_product():
    """T√¨m ki·∫øm s·∫£n ph·∫©m theo m√£ s·∫£n ph·∫©m"""
    print("\n--- T√åM KI·∫æM S·∫¢N PH·∫®M ---")
    
    if not products:
        print(" Danh s√°ch s·∫£n ph·∫©m tr·ªëng!")
        return
    
    product_id = input("Nh·∫≠p m√£ s·∫£n ph·∫©m c·∫ßn t√¨m: ").strip()
    index = find_product_by_id(product_id)
    
    if index == -1:
        print(f" Kh√¥ng t√¨m th·∫•y s·∫£n ph·∫©m c√≥ m√£ '{product_id}'!")
    else:
        product = products[index]
        print(f"\n T√¨m th·∫•y s·∫£n ph·∫©m:")
        print(f"M√£ s·∫£n ph·∫©m: {product[0]}")
        print(f"T√™n s·∫£n ph·∫©m: {product[1]}")
        print(f"S·ªë l∆∞·ª£ng: {product[2]}")
        print(f"Gi√°: {product[3]:,.0f} VND")

def update_product():
    """C·∫≠p nh·∫≠t th√¥ng tin s·∫£n ph·∫©m"""
    print("\n--- C·∫¨P NH·∫¨T TH√îNG TIN S·∫¢N PH·∫®M ---")
    
    if not products:
        print(" Danh s√°ch s·∫£n ph·∫©m tr·ªëng!")
        return
    
    product_id = input("Nh·∫≠p m√£ s·∫£n ph·∫©m c·∫ßn c·∫≠p nh·∫≠t: ").strip()
    index = find_product_by_id(product_id)
    
    if index == -1:
        print(f" Kh√¥ng t√¨m th·∫•y s·∫£n ph·∫©m c√≥ m√£ '{product_id}'!")
        return
    
    # Hi·ªÉn th·ªã th√¥ng tin hi·ªán t·∫°i
    current_product = products[index]
    print(f"\nTh√¥ng tin hi·ªán t·∫°i:")
    print(f"M√£ s·∫£n ph·∫©m: {current_product[0]}")
    print(f"T√™n s·∫£n ph·∫©m: {current_product[1]}")
    print(f"S·ªë l∆∞·ª£ng: {current_product[2]}")
    print(f"Gi√°: {current_product[3]:,.0f} VND")
    
    # Nh·∫≠p th√¥ng tin m·ªõi
    print("\nNh·∫≠p th√¥ng tin m·ªõi (Enter ƒë·ªÉ gi·ªØ nguy√™n):")
    
    new_name = input(f"T√™n s·∫£n ph·∫©m [{current_product[1]}]: ").strip()
    if not new_name:
        new_name = current_product[1]
    
    try:
        quantity_input = input(f"S·ªë l∆∞·ª£ng [{current_product[2]}]: ").strip()
        new_quantity = int(quantity_input) if quantity_input else current_product[2]
        
        price_input = input(f"Gi√° [{current_product[3]}]: ").strip()
        new_price = float(price_input) if price_input else current_product[3]
        
        if new_quantity < 0 or new_price < 0:
            print(" L·ªói: S·ªë l∆∞·ª£ng v√† gi√° kh√¥ng th·ªÉ √¢m!")
            return
        
        # C·∫≠p nh·∫≠t s·∫£n ph·∫©m
        products[index] = (product_id, new_name, new_quantity, new_price)
        print(f" ƒê√£ c·∫≠p nh·∫≠t s·∫£n ph·∫©m '{product_id}' th√†nh c√¥ng!")
        
    except ValueError:
        print(" L·ªói: Vui l√≤ng nh·∫≠p s·ªë h·ª£p l·ªá cho s·ªë l∆∞·ª£ng v√† gi√°!")

def delete_product():
    """X√≥a s·∫£n ph·∫©m theo m√£ s·∫£n ph·∫©m"""
    print("\n--- X√ìA S·∫¢N PH·∫®M ---")
    
    if not products:
        print(" Danh s√°ch s·∫£n ph·∫©m tr·ªëng!")
        return
    
    product_id = input("Nh·∫≠p m√£ s·∫£n ph·∫©m c·∫ßn x√≥a: ").strip()
    index = find_product_by_id(product_id)
    
    if index == -1:
        print(f" Kh√¥ng t√¨m th·∫•y s·∫£n ph·∫©m c√≥ m√£ '{product_id}'!")
        return
    
    # Hi·ªÉn th·ªã th√¥ng tin s·∫£n ph·∫©m s·∫Ω x√≥a
    product = products[index]
    print(f"\nS·∫£n ph·∫©m s·∫Ω b·ªã x√≥a:")
    print(f"M√£: {product[0]} | T√™n: {product[1]} | SL: {product[2]} | Gi√°: {product[3]:,.0f} VND")
    
    # X√°c nh·∫≠n x√≥a
    confirm = input("B·∫°n c√≥ ch·∫Øc ch·∫Øn mu·ªën x√≥a? (y/n): ").strip().lower()
    if confirm in ['y', 'yes']:
        products.pop(index)
        print(f" ƒê√£ x√≥a s·∫£n ph·∫©m '{product[1]}' th√†nh c√¥ng!")
    else:
        print(" H·ªßy thao t√°c x√≥a!")

def main():
    """H√†m ch√≠nh ƒëi·ªÅu khi·ªÉn ch∆∞∆°ng tr√¨nh"""
    print("  CH√ÄO M·ª™NG ƒê√äN H·ªÜ TH·ªêNG QU·∫¢N L√ù S·∫¢N PH·∫®M C·ª¨A H√ÄNG ")
    
    while True:
        display_menu()
        
        try:
            choice = input("Vui l√≤ng ch·ªçn ch·ª©c nƒÉng (0-5): ").strip()
            
            if choice == '1':
                add_product()
            elif choice == '2':
                display_products()
            elif choice == '3':
                search_product()
            elif choice == '4':
                update_product()
            elif choice == '5':
                delete_product()
            elif choice == '0':
                print("\n C·∫£m ∆°n b·∫°n ƒë√£ s·ª≠ d·ª•ng h·ªá th·ªëng!")
                print(" T·∫°m bi·ªát!")
                break
            else:
                print(" L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá! Vui l√≤ng ch·ªçn t·ª´ 0-5.")
                
        except KeyboardInterrupt:
            print("\n\n C·∫£m ∆°n b·∫°n ƒë√£ s·ª≠ d·ª•ng h·ªá th·ªëng!")
            print(" T·∫°m bi·ªát!")
            break
        except Exception as e:
            print(f" ƒê√£ x·∫£y ra l·ªói: {e}")
        
        # T·∫°m d·ª´ng ƒë·ªÉ ng∆∞·ªùi d√πng ƒë·ªçc th√¥ng b√°o
        input("\nNh·∫•n Enter ƒë·ªÉ ti·∫øp t·ª•c...")

# Kh·ªüi ch·∫°y ch∆∞∆°ng tr√¨nh
if __name__ == "__main__":
    main()
```
