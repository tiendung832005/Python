```python
# Bài 7:
products = {"A": 120, "B": 80, "C": 300, "D": 50}
print("Danh sách sản phẩm ban đầu:", products)

products_to_remove = [key for key, value in products.items() if value < 100]
for key in products_to_remove:
    del products[key]

print("Danh sách sản phẩm sau khi xóa (giá >= 100):", products)

print("Danh sách sản phẩm sau khi thêm sản phẩm E:", products)

# Thêm sản phẩm mới
products["E"] = 250  # Thêm sản phẩm E với giá 250
print("Danh sách sản phẩm sau khi thêm sản phẩm E:", products)

product_count = len(products)
print(f"Số lượng sản phẩm còn lại: {product_count}")


print("\nThông tin chi tiết:")print(f"- Tổng giá trị các sản phẩm còn lại: {sum(products.values())}")

print(f"- Sản phẩm đã xóa: {products_to_remove}")print(f"- Sản phẩm còn lại: {list(products.keys())}")
```
