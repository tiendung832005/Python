```python
import matplotlib.pyplot as plt

# Chuẩn bị về dữ liệu
months = [1,2,3]
sales = [100, 200, 300]

#khai báo figure và axes
fig, ax = plt.subplots()

# vẽ biểu đồ từ các dữ liệu trên
ax.plot(months, sales)


# Đặt nhãn, tiêu đề cho biểu đồ
ax.set_title('Biểu đồ doanh thu tháng 11-2025')
ax.set_xlabel('Thời gian')
ax.set_ylabel('Doanh thu')

#hiển thị
plt.show()
```
