```python
# Bài 1: Vẽ đồ thị cơ bản
import matplotlib.pyplot as plt
import numpy as np

# Tạo mảng x từ 0 đến 10
x = np.linspace(0, 10, 50)
# Tạo đường y = x²
y = x**2

plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.title("Biểu đồ đầu tay của [Tên bạn]", fontsize=14, fontweight='bold')
plt.xlabel("Trục x", fontsize=12)
plt.ylabel("Trục y", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Bài 2: Vẽ đồ thị hàm bậc ba
import matplotlib.pyplot as plt
import numpy as np

# Tạo x từ -5 đến 5
x = np.linspace(-5, 5, 100)
# Vẽ y = x³
y = x**3

plt.figure(figsize=(8, 6))
plt.plot(x, y, 'r-', linewidth=3)
plt.title("Đồ thị hàm bậc ba y = x³", fontsize=14, fontweight='bold')
plt.xlabel("Giá trị x", fontsize=12)
plt.ylabel("Giá trị y", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Bài 3: Vẽ đường thẳng
import matplotlib.pyplot as plt
import numpy as np

# Tạo mảng x từ 0 đến 10 (20 điểm)
x = np.linspace(0, 10, 20)
# Tính y = 2x + 3
y = 2*x + 3

plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=3)
plt.title("Đường thẳng y = 2x + 3", fontsize=14, fontweight='bold')
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.grid(True, alpha=0.7)
plt.show()
```

```python
# Bài 4: So sánh sin và cos
import matplotlib.pyplot as plt
import numpy as np

# Tạo mảng x từ 0 đến 2π với 200 điểm
x = np.linspace(0, 2*np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, 'b-', linewidth=2, label='sin(x)')
plt.plot(x, y2, 'r--', linewidth=2, label='cos(x)')
plt.title("So sánh sin(x) và cos(x)", fontsize=14, fontweight='bold')
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.7)
plt.show()
```

```python
# Bài 5: Giao điểm hai đường thẳng
import matplotlib.pyplot as plt
import numpy as np

# Tạo x từ 0 đến 10
x = np.linspace(0, 10, 100)
y1 = 2*x + 1  # Đường 1
y2 = -x + 10  # Đường 2

plt.figure(figsize=(8, 6))
plt.plot(x, y1, 'b-', linewidth=2, label='Đường 1')
plt.plot(x, y2, 'orange', linewidth=2, label='Đường 2')
plt.title("Giao điểm của hai đường thẳng", fontsize=14, fontweight='bold')
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, alpha=0.7)
plt.show()
```

```python
# Bài 6: Tăng trưởng dân số Việt Nam
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
population = [92, 93, 94, 95, 96, 97, 98, 99, 100, 101]

plt.figure(figsize=(10, 6))
plt.plot(years, population, 'b-o', linewidth=2, markersize=6)
plt.title("Tăng trưởng dân số Việt Nam 2015–2024", fontsize=14, fontweight='bold')
plt.xlabel("Năm", fontsize=12)
plt.ylabel("Dân số (triệu người)", fontsize=12)
plt.grid(True, alpha=0.7)
plt.legend(['Dân số'], loc='lower right')
plt.tight_layout()
plt.show()
```

```python
# Bài 7: Doanh số bán hàng
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
products = ['Áo', 'Quần', 'Giày', 'Túi', 'Mũ', 'Tất']
sales = [150, 120, 200, 90, 65, 80]

plt.figure(figsize=(10, 6))
bars = plt.bar(products, sales, color='green', edgecolor='black', linewidth=1.5)

# In giá trị doanh số lên đầu mỗi cột
for bar, sale in zip(bars, sales):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2, 
             f'{sale}', ha='center', va='bottom', color='white', fontweight='bold')

plt.title("Doanh số bán hàng tháng 10/2025", fontsize=14, fontweight='bold')
plt.xlabel("Sản phẩm", fontsize=12)
plt.ylabel("Doanh số (triệu)", fontsize=12)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

```python
# Bài 8: Nhiệt độ và lượng mưa Hà Nội
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu
months = list(range(1, 13))
temperature = [20, 22, 25, 28, 30, 32, 31, 30, 28, 26, 23, 21]
rainfall = [10, 15, 30, 120, 200, 250, 280, 220, 150, 80, 30, 15]

plt.figure(figsize=(12, 6))

# Vẽ đường nhiệt độ
plt.plot(months, temperature, 'r-', linewidth=2, label='Nhiệt độ (°C)')

# Vẽ điểm lượng mưa
plt.scatter(months, rainfall, color='blue', s=80, label='Lượng mưa (mm)', alpha=0.7)

plt.title("Nhiệt độ và lượng mưa Hà Nội 2025", fontsize=14, fontweight='bold')
plt.xlabel("Tháng", fontsize=12)
plt.ylabel("Giá trị", fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, alpha=0.7)
plt.xticks(months)
plt.tight_layout()
plt.show()
```
